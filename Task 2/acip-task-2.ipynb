```python
class TrainJourney:
    def __init__(self, departure_time, return_time, max_passengers, ticket_price):
        self.departure_time = departure_time
        self.return_time = return_time
        self.max_passengers = max_passengers
        self.ticket_price = ticket_price
        self.available_tickets = max_passengers
        self.total_passengers = 0
        self.total_revenue = 0

    def book_ticket(self, num_passengers):
        if num_passengers <= self.available_tickets:
            self.available_tickets -= num_passengers
            self.total_passengers += num_passengers
            self.total_revenue += num_passengers * self.ticket_price
            return True
        else:
            return False

    def is_full(self):
        return self.available_tickets == 0

    def display_status(self):
        if self.is_full():
            return "Closed"
        else:
            return f"Available Tickets: {self.available_tickets}, Total Passengers: {self.total_passengers}, Total Revenue: ${self.total_revenue}"


# Initialize train journeys
journeys = [
    TrainJourney("09:00", "10:00", 80, 25),
    TrainJourney("11:00", "12:00", 80, 25),
    TrainJourney("13:00", "14:00", 80, 25),
    TrainJourney("15:00", "16:00", 80, 25),
]

for journey in journeys:
    print(f"\nJourney {journey.departure_time} - {journey.return_time}: {journey.display_status()}")

# User input for booking tickets
journey_index = int(input("\nEnter the journey index (0-3) to book tickets: "))
num_passengers = int(input("Enter the number of passengers: "))

# Booking logic
if 0 <= journey_index < len(journeys):
    if journeys[journey_index].book_ticket(num_passengers):
        print("Tickets booked successfully!")
    else:
        print("Tickets not available.")
else:
    print("Invalid journey index.")
```

```python
class TrainJourney:
    def __init__(self, departure_time, return_time, ticket_price, total_tickets):
        self.departure_time = departure_time
        self.return_time = return_time
        self.ticket_price = ticket_price
        self.total_tickets = total_tickets
        self.available_tickets = total_tickets
        self.total_passengers = 0
        self.total_money = 0

    def display_status(self):
        if self.available_tickets > 0:
            return f"{self.available_tickets} tickets available"
        else:
            return "Closed"

    def book_ticket(self, num_tickets):
        if self.available_tickets >= num_tickets:
            self.available_tickets -= num_tickets
            return True
        else:
            return False

    def purchase_tickets(self, num_passengers):
        if self.available_tickets >= num_passengers:
            self.available_tickets -= num_passengers
            total_price = num_passengers * self.ticket_price

            # Check for group discount
            if num_passengers >= 10:
                free_tickets = num_passengers // 10
                total_price -= free_tickets * self.ticket_price

            self.total_passengers += num_passengers
            self.total_money += total_price

            return total_price
        else:
            return 0  # Tickets not available

# Initialize train journeys
journeys = [
    TrainJourney("09:00", "10:00", 25.00, 80),
    TrainJourney("11:00", "12:00", 25.00, 80),
    TrainJourney("13:00", "14:00", 25.00, 80),
    TrainJourney("15:00", "16:00", 25.00, 80)
]

# Display initial status for each journey
for journey in journeys:
    print(f"\nJourney {journey.departure_time} - {journey.return_time}: {journey.display_status()}")

# User input for purchasing tickets
while True:
    journey_index = int(input("\nEnter the journey index (0-3) to purchase tickets (or enter -1 to stop): "))
    
    if journey_index == -1:
        break

    num_passengers = int(input("Enter the number of passengers: "))

    # Purchase logic
    if 0 <= journey_index < len(journeys):
        total_price = journeys[journey_index].purchase_tickets(num_passengers)

        if total_price > 0:
            print(f"Tickets purchased successfully! Total price: ${total_price:.2f}")
        else:
            print("Tickets not available.")
    else:
        print("Invalid journey index.")

    # Display updated status for the chosen journey
    print(f"Journey {journeys[journey_index].departure_time} - {journeys[journey_index].return_time}: "
          f"{journeys[journey_index].display_status()}")

# Display final status for each journey
for journey in journeys:
    print(f"\nJourney {journey.departure_time} - {journey.return_time}: {journey.display_status()}")

# Display number of passengers and total money for each journey
print("\nJourney Statistics:")
for journey in journeys:
    print(f"Journey {journey.departure_time} - {journey.return_time}: {journey.total_passengers} passengers, "
          f"Total money: ${journey.total_money:.2f}")

# Calculate and display total number of passengers and total money for the day
total_passengers_day = sum(journey.total_passengers for journey in journeys)
total_money_day = sum(journey.total_money for journey in journeys)
print(f"\nTotal Passengers for the Day: {total_passengers_day} passengers")
print(f"Total Money for the Day: ${total_money_day:.2f}")

# Find and display the journey with the most passengers
most_passengers_journey = max(journeys, key=lambda x: x.total_passengers)
print(f"\nJourney with the Most Passengers: "
      f"{most_passengers_journey.departure_time} - {most_passengers_journey.return_time}, "
      f"{most_passengers_journey.total_passengers} passengers")

```
